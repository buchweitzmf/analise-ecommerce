```python
df2 = pd.read_csv('Amazon Sale Report.csv', encoding = 'latin1')
df2.head()
```

```python
df2.shape
```

```python
df2.dtypes
```

```python
df2['ship-state'] = df2['ship-state'].str.upper()
```

```python
df2.drop(
    labels=['Unnamed: 22'],
    axis=1,
    inplace=True
)
```

```python
df2['Date'] = pd.to_datetime(df2['Date'], format='%m-%d-%y')
```

```python
df2['Date'].head()
```

```python
df2.isnull().sum()
```

```python
df2.drop(
    labels=['fulfilled-by'],
    axis=1,
    inplace=True
)
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt
```

```python
print(df2.groupby('Category')['Qty'].sum().sort_values(ascending=False))
```

```python
vendas_categoria = df2.groupby('Category')['Qty'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
vendas_categoria.plot(kind='bar')
plt.title('Vendas por categoria')
plt.xlabel('Categoria')
plt.ylabel('Quantidade')
plt.tight_layout()
plt.show()
```

```python
df2.groupby(df2['Date'].dt.to_period('M'))['Amount'].sum().sort_index()
```

```python
evolucao_vendas = df2.groupby(df2['Date'].dt.to_period('M'))['Amount'].sum().sort_index()
plt.figure(figsize=(10, 6))
evolucao_vendas.plot(kind='line')
plt.title('EvoluÃ§Ã£o das Vendas')
plt.xlabel('Data')
plt.ylabel('Total de Vendas')
plt.tight_layout()
plt.show()
```

```python
df2.head()
```

```python
df2.groupby(df2['ship-state'])['Amount'].sum().sort_values(ascending=False)
```

```python
estados_mais_compram = df2.groupby(df2['ship-state'])['Amount'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
estados_mais_compram.plot(kind='bar')
plt.title('Estados que mais compram')
plt.xlabel('Estados')
plt.ylabel('Total em compras')
plt.tight_layout()
plt.show()
```
